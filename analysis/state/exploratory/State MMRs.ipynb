This notebook just explores and plots the various MMRs by state

```python
# Imports
import os
import psycopg2 as dbconnect
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import plotly.plotly as py
#from plotly.graph_objs import *
import plotly.tools as tls
```

```python
# Setup other vars
export_dir = '/Users/Nicke/desktop/dssg/mexico/working_area'
db = dbconnect.connect(host='',
                             database='mexico',
                             user='',
                             password='')
```

```python
# Import births & maternal fatalities by state & year
births = pd.read_sql("""
SELECT count(*) as births, 
    mother_res_state,
    extract(year from child_dob) as birth_year
FROM births_data
GROUP BY mother_res_state, birth_year
;""", db)

maternal_deaths = pd.read_sql("""
SELECT count(*) as maternal_deaths,
    residence_state,
    extract(year from date_of_death) as death_year
FROM maternal_mortality_data
GROUP BY residence_state, death_year
;""", db)

state_info = pd.read_sql("""
SELECT cvegeo as state,
    nombre as state_name
FROM census.national_estatal
;""", db)
```

```python
# Some data munging
births.rename(columns={'mother_res_state':'state', 'birth_year': 'year'}, inplace=True)
maternal_deaths.rename(columns={'residence_state':'state', 'death_year':'year'}, inplace=True)

data = pd.merge(births, maternal_deaths, 
                how='outer',
                on = ['state', 'year'])
data.fillna(0, inplace=True)
data = pd.merge(data, state_info, how='left', on='state')
data = data[(data.year>=2008) & (data.year<=2012)]
```

```python
# Just look at 2012 for simplicity
data = data[data.year==2012]
```

```python
data['mmr'] = (data.maternal_deaths / data.births) * 100000
data['pct_deaths'] = data.maternal_deaths / data.maternal_deaths.sum() * 100
```

```python
data[~data.state_name.isnull()].set_index('state_name').mmr.plot(kind='bar')
plt.title('MMR by State')
plt.ylabel('MMR')
plt.xlabel('')
plt.show()
data[~data.state_name.isnull()].set_index('state_name')
```

```python
data[~data.state_name.isnull()].set_index('state_name').pct_deaths.plot(kind='bar')
plt.title('Percent of Deaths by State')
plt.ylabel('Percent of Deaths (%)')
plt.xlabel('')
```

```python
#### Compare these MMRs from births datasets with MMRs from CONAPO datasets
conapo_births = pd.read_sql("""
SELECT CONCAT(CASE WHEN state_id::int < 10 THEN '0' ELSE '' END, state_id) as state_id,
    year,
    births as conapo_births
FROM conapo.sociodemographicvariables
""",db)
data = pd.merge(data.drop('mmr', axis=1), conapo_births, left_on=['state', 'year'], right_on=['state_id', 'year'], how='left')
```

```python
# Just look at 2012 for simplicity
data = data[data.year==2012]
data['mmr'] = (data.maternal_deaths / data.conapo_births) * 100000
data['pct_deaths'] = data.maternal_deaths / data.maternal_deaths.sum() * 100
```

```python
data[~data.state_name.isnull()].set_index('state_name').mmr.plot(kind='bar')
plt.title('MMR by State')
plt.ylabel('MMR')
plt.xlabel('')
plt.show()
data[~data.state_name.isnull()].set_index('state_name')
```
