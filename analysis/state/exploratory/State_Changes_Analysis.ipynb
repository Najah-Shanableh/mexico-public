### Analysis to determine the states that have seen the greatest changes in MMR over a given time period


```python
# Imports
import os
import psycopg2 as dbconnect
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import plotly.plotly as py
#from plotly.graph_objs import *
import plotly.tools as tls
```

```python
# Setup other vars
export_dir = '/Users/Nicke/desktop/dssg/mexico/working_area'
db = dbconnect.connect(host='',
                             database='mexico',
                             user='',
                             password='')
```

```python
# Import births & maternal fatalities by state & year
births = pd.read_sql("""
SELECT count(*) as births, 
    mother_res_state,
    extract(year from child_dob) as birth_year
FROM births_data
GROUP BY mother_res_state, birth_year
;""", db)

maternal_deaths = pd.read_sql("""
SELECT count(*) as maternal_deaths,
    residence_state,
    extract(year from date_of_death) as death_year
FROM maternal_mortality_data
GROUP BY residence_state, death_year
;""", db)

state_info = pd.read_sql("""
SELECT cvegeo as state,
    nombre as state_name
FROM census.national_estatal
;""", db)
```

```python
# Some data munging
births.rename(columns={'mother_res_state':'state', 'birth_year': 'year'}, inplace=True)
maternal_deaths.rename(columns={'residence_state':'state', 'death_year':'year'}, inplace=True)

data = pd.merge(births, maternal_deaths, 
                how='outer',
                on = ['state', 'year'])
data.maternal_deaths.fillna(0, inplace=True)
```

### Analysis by Year
Notes: Big improvement in MMR from 2009 to 2010, but it hits an inflection point and levels off

```python
# Plot absolute MMR over time
year_data = data.groupby('year').sum().reset_index()
year_data = year_data[(year_data.year >= 2008) & (year_data.year <= 2012)]
year_data['mmr'] = year_data.maternal_deaths/year_data.births * 100000

plt.xticks(year_data.year, [str(int(i)) for i in year_data.year])
plt.plot(year_data.year, year_data.mmr)
plt.show()
```

```python
# MMR Change over time
year_data['mmr_change'] = float('NaN')
year_data['pct_mmr_change'] = float('NaN')
for year in year_data.year:
    try:    
        year_data['mmr_change'][year_data.year==year] = year_data.mmr[year_data.year==year].values[0] - year_data.mmr[year_data.year==year-1].values[0]
        year_data['pct_mmr_change'][year_data.year==year] = (year_data['mmr_change'][year_data.year==year].values[0] / year_data['mmr'][year_data.year==year-1].values[0])*100
    except:
        pass
year_data
```

```python
# Plot percent change over time
plt.xticks(year_data.year, [str(int(i)) for i in year_data.year])
plt.plot(year_data.year, year_data.pct_mmr_change)
plt.show()
```

### Analysis by state & year
Notes: Some really good places and some really bad places
- Really bad places: Aguascalientes(+57%), Sonora (+53%), Tlaxcala (+140%)
- Really good places: Chiapas(-45%), Baja California (-40%), Oaxaca (-52%)
These places don't look spatially correlated so we might need to look deeper the factors affecting each.
All places that increased drastically were below average in MMR in 2008, all that showed good improvements were above average especially Chiapas and Oaxaca

```python
# Plot by state and year
state_data = data[(data.year >= 2008) & (data.year <= 2012)]
state_data['mmr'] = state_data.maternal_deaths/state_data.births * 100000
state_data.sort(columns=['state', 'year'], inplace=True)
state_data = pd.merge(state_data, state_info, how='left', on='state')
state_data.state_name.fillna('unknown', inplace=True)

for key, group in state_data.groupby(['state_name', 'state']):
    plt.xticks(group.year, [str(int(i)) for i in group.year])
    plt.title(key[0] + ", code: " + key[1] + ", births: " + str(int(group.births.sum())) + ", deaths: " + str(int(group.maternal_deaths.sum())))
    plt.xlabel('year')
    plt.ylabel('mmr')
    plt.plot(group.year, group.mmr)
    plt.show()
```

```python
# Calculate percent MMR change from 2008 to 2012 by state
melted_data = data
melted_data = melted_data[(melted_data.year >= 2008) & (melted_data.year <= 2012)]
melted_data['mmr'] = melted_data.maternal_deaths / melted_data.births * 100000

summary_data = pd.DataFrame()
for state, group in melted_data.groupby('state'):
    try:
        mmr_change = group.mmr[group.year==2012].values[0] - group.mmr[group.year==2008].values[0]
        pct_mmr_change = mmr_change / group.mmr[group.year==2008].values[0] * 100
        summary_data = summary_data.append({'state': state,
                                            'mmr_change': mmr_change,
                                            'pct_mmr_change': pct_mmr_change,
                                            'mmr_2008': group.mmr[group.year==2008].values[0],
                                            'mmr_2012': group.mmr[group.year==2012].values[0],
                                            'births_2008': group.births[group.year==2008].values[0],
                                            'births_2012': group.births[group.year==2012].values[0],
                                            'maternal_deaths_2008': group.maternal_deaths[group.year==2008].values[0],
                                            'maternal_deaths_2012': group.maternal_deaths[group.year==2012].values[0]
                                            }, ignore_index=True)
    except:
        pass
summary_data = pd.merge(summary_data, state_info, how='left', on='state')
from statsmodels.stats import proportion
summary_data['chi_sq'] = summary_data.apply(lambda row: proportion.proportions_chisquare([row.maternal_deaths_2008, row.maternal_deaths_2012], [row.births_2008, row.births_2012])[1], axis=1)
#summary_data.to_csv(os.path.join(export_dir, 'state_summary.csv'), index=False)
summary_data
```

```python
### Now use CONAPO Data for MMR calculations
conapo_births = pd.read_sql("""
SELECT CONCAT(CASE WHEN state_id::int < 10 THEN '0' ELSE '' END, state_id) as state_id,
    year,
    births as births
FROM conapo.sociodemographicvariables
""",db)
data = pd.merge(births, maternal_deaths, 
                how='outer',
                on = ['state', 'year'])
data.maternal_deaths.fillna(0, inplace=True)
conapo_data = pd.merge(data.drop('births', axis=1), conapo_births, left_on=['state', 'year'], right_on=['state_id', 'year'], how='left')
conapo_data = conapo_data[(conapo_data.year >= 2002) & (conapo_data.year <= 2012)]
conapo_data = conapo_data[~conapo_data.state.isin(['00','33','34','35','36'])]
conapo_data['mmr'] = conapo_data.maternal_deaths / conapo_data.births * 100000
```

```python
def summarize_conapo_change(data, start_year, end_year):
    summary_data = pd.DataFrame()
    for state, group in data.groupby('state'):
        try:
            mmr_change = group.mmr[group.year==end_year].values[0] - group.mmr[group.year==start_year].values[0]
            pct_mmr_change = mmr_change / group.mmr[group.year==start_year].values[0] * 100
            summary_data = summary_data.append({'state': state,
                                                'mmr_change': mmr_change,
                                                'pct_mmr_change': pct_mmr_change,
                                                'mmr_'+str(start_year): group.mmr[group.year==start_year].values[0],
                                                'mmr_'+str(end_year): group.mmr[group.year==end_year].values[0],
                                                'births_'+str(start_year): group.births[group.year==start_year].values[0],
                                                'births_'+str(end_year): group.births[group.year==end_year].values[0],
                                                'maternal_deaths_'+str(start_year): group.maternal_deaths[group.year==start_year].values[0],
                                                'maternal_deaths_'+str(end_year): group.maternal_deaths[group.year==end_year].values[0]
                                                }, ignore_index=True)
        except:
            pass
    summary_data = pd.merge(summary_data, state_info, how='left', on='state')
    from statsmodels.stats import proportion
    summary_data['chi_sq'] = summary_data.apply(lambda row: proportion.proportions_chisquare([row['maternal_deaths_'+str(start_year)], row['maternal_deaths_'+str(end_year)]], [row['births_'+str(start_year)], row['births_'+str(end_year)]])[1], axis=1)
    return summary_data
```
