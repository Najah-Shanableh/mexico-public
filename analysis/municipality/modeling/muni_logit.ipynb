```python
def to_percentiles(a, q=range(90,-10,-10)):
    percentiles = np.percentile(a=a, q=q)
    print percentiles
    return a.apply(lambda x: len(q) - next(i for i,p in enumerate(percentiles) if x >= p))

def decile_plot(a, x, y='mmrate', z=None):
    z = z or np.mean(a[y])
    print z
    s = pd.DataFrame({y: a[y], x:to_percentiles(a[x])}).groupby(x)\
        .apply(lambda d: sum(d[y] > z)*1.0/len(d))
    df = pd.Series(s)
    ax =  df.plot(kind='bar')
    ax.set_ylabel('% above mean ' + y)
    return ax
```

```python
urban_data = subset_data[(subset_data.urban_births > 0) & (subset_data.urban_pop_birth > 0)]
mmbinary = urban_data.urban_mmbinary
data = urban_data.drop(['nom_ent', 'nom_mun', 'municipality', 'ent', 'deaths', 'mmrate', 'mmbinary', \
    'urban_mmbinary','urban_mmrate', 'rural_mmrate', 'urban_deaths', 'rural_deaths', \
    'urban_pop_fem', 'urban_pop', 'urban_pop_birth', 'pob_tot'], axis=1)
data = data.fillna(0)
```

```python
scaler = preprocessing.StandardScaler()
scaled_data = scaler.fit_transform(data)
scaled_data = pd.DataFrame(scaled_data)
scaled_data.columns = data.columns
scaled_data.index = data.index
```

```python
scaled_data.basic_margin = scaled_data.margination * scaled_data.basico_distance
```

```python
kf = cross_validation.KFold(n=len(scaled_data), n_folds=5, shuffle=True, random_state=None)
models = []
predictions = {}
for train_index, test_index in kf:
    print 'training'
    m = {}
    m['X_train'], m['X_test'] = scaled_data.irow(train_index), scaled_data.irow(test_index)
    m['y_train'], m['y_test'] = mmbinary.irow(train_index), mmbinary.irow(test_index)
    
    m['model'] =  linear_model.LogisticRegression()#penalty='l1', C=1)
    m['model'].fit(m['X_train'], m['y_train'])
    
    print 'testing'
    m['predict'] = m['model'].predict(m['X_test'])
    m['predict_proba'] = m['model'].predict_proba(m['X_test'])[:,1]
    models.append(m)
    #print(sum(y_test*predictions)/sum(y_test).astype('float'))
    #print sum(y_test*predictions)/sum(predictions).astype('float')
```

```python
np.mean([metrics.roc_auc_score(y_true=m['y_test'].values, y_score = m['predict_proba']) for m in models])
```
