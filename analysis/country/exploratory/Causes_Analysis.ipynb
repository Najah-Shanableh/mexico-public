# Analysis of Causes of Death for Mexico

### Key Findings
- Only about 1-1.5% of maternal deaths happen due to causes outside of pregnancy
- At a high level, the following 3 are the primary causes of death, comprising around 20-25% of deaths each:
    1. Complications related to labor
    1. Oedema, proteinuria, and hypertensive disorders (related to pregnancy)
    1. Other obstetric conditions
- From 2002-2012, there have been large declines in the proportions of deaths relating to labor complications and oedema, proteinuria, and hypertensive disorders, but a large increase in "other" conditions
- Unfortunately, our data does not have high enough resolution to determine what's going on in the "other" disorders

### Key Questions
- Are doctors just checking off "other"?
- Are there higher-resolution cause information available?
- Are there types of diseases that doctors usually just check this info off for?

```python
### Setup
import os
import psycopg2 as dbconnect
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import csv
from statsmodels.stats import proportion

export_dir = '/Users/Nicke/desktop/dssg/mexico/working_area'
if not os.path.exists(export_dir):
    os.makedirs(export_dir)
    
### Data connection
db = dbconnect.connect(host='',
                             database='mexico',
                             user='',
                             password='')
```

```python
### Import Data
data = pd.read_sql("""
SELECT maternal_mortality_data.*,
    icd10_3.code3_description,
    icd10_3.block_description,
    icd10_3.chapter_description,
    icd10_4.code4_description
FROM maternal_mortality_data
LEFT JOIN icd10_3
USING (icd_10_3)
LEFT JOIN icd10_4
USING (icd_10_4)
""", db)
```

```python
# Check to see if there were any non matches for icd3's
data[(data.code3_description.isnull()) | (data.block_description.isnull()) | (data.chapter_description.isnull())].empty
```

```python
# Look at only years that're supposed to be there
data = data[(data.dod_year >= 2002) & (data.dod_year <= 2012)]
```

```python
# Look at chapter level info
chapter_freq = 100. * data.chapter_description.value_counts() / len(data.chapter_description)
blah = chapter_freq.plot(kind='bar')
plt.ylabel('Percent of Deaths')
plt.xlabel('ICD10 Chapter')
plt.show()
pd.DataFrame({'percent': chapter_freq})
```

```python
# Look at block level info
block_freq = 100. * data.block_description.value_counts() / len(data.block_description)
block_freq.plot(kind='bar')
plt.ylabel('Percent of Deaths')
plt.show()
pd.DataFrame({'percent':block_freq})
```

```python
# Look at change over time
year_change = pd.DataFrame({'reasons': data.block_description.unique()})
sorted_years = data.dod_year.unique()
sorted_years.sort()
for year in sorted_years:
    year_data = data[data.dod_year == year]
    year_block_freq = 100. * year_data.block_description.value_counts() / len(year_data.block_description)
    temp = pd.DataFrame({str(int(year)): year_block_freq})
    year_change = pd.merge(year_change, temp, left_on='reasons', right_index=True, how='left')
year_change.plot()
plt.ylabel('Percent of Deaths')
plt.show()
```

```python
# Let's group this shit by year
year_groups = pd.DataFrame({'year': range(2002, 2013), 'year_group': ['02-04', '02-04', '02-04',
                                                                       '05-08', '05-08', '05-08', '05-08',
                                                                       '09-12', '09-12','09-12','09-12']})

data_groups = pd.merge(data, year_groups, left_on='dod_year', right_on='year', how='left')
sorted_groups = data_groups.year_group.unique()
sorted_groups.sort()
group_change = pd.DataFrame({'reasons': data_groups.block_description.unique()})
for group in sorted_groups:
    group_data = data_groups[data_groups.year_group == group]
    group_block_freq = 100. * group_data.block_description.value_counts() / len(group_data.block_description)
    temp = pd.DataFrame({str(group): group_block_freq})
    group_change = pd.merge(group_change, temp, left_on='reasons', right_index=True, how='left')
group_change.set_index('reasons').plot(kind='barh')
plt.title('Causes of Death over Time')
plt.xlabel('Percent of Deaths')
plt.ylabel('')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Let's look more into what these other complications are
other = data[data.block_description=='Other obstetric conditions, not elsewhere classified']
other = other[['chapter_description', 'block_description', 'code3_description', 'code4_description']]
other.code4_description[other.code4_description.isnull()] = 'unknown'
other_pct = 100. * other.code4_description.value_counts() / len(other)
other_pct.plot(kind='bar')
plt.title("Distribution of 'other' causes")
plt.ylabel("Percentage of Deaths")

pd.DataFrame({'percent': other_pct})
```

```python
### Are there specific causes that plague Mexico City?
state_info = pd.read_sql("""
SELECT cvegeo as state,
    nombre as state_name
FROM census.national_estatal
;""", db)

data = pd.merge(data, state_info, left_on='residence_state', right_on='state', how='left')
```

```python
# only look at data for the last few years
mc_data = data[(data.dod_year >= 2008) & (data.dod_year <= 2012) & (data.residence_state=='15')]
```

```python
# Look at block level info
mc_block_freq = 100. * mc_data.block_description.value_counts() / len(mc_data.block_description)
mc_block_freq.plot(kind='bar')
plt.ylabel('Percent of Deaths')
plt.show()
pd.DataFrame({'percent':mc_block_freq})
```

```python
# Look at code level info
mc_code_freq = 100. * mc_data.code4_description.value_counts() / len(mc_data.code4_description)
mc_code_freq.plot(kind='bar')
plt.ylabel('Percent of Deaths')
plt.show()
pd.DataFrame({'percent':mc_code_freq})
```
