## Analysis to determine the differences between urban and rural MMRs

```python
### Setup
import os
import psycopg2 as dbconnect
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import csv
from statsmodels.stats import proportion
```

```python
### Data connection
db = dbconnect.connect(host='',
                             database='mexico',
                             user='',
                             password='')
```

```python
### Import births & maternal fatalities by state & year, and import localities info
births = pd.read_sql("""
SELECT count(*) as births, 
    mother_res_state||mother_res_muni||mother_res_loc as locality_code,
    extract(year from child_dob) as birth_year
FROM births_data
GROUP BY locality_code, birth_year
;""", db)

maternal_deaths = pd.read_sql("""
SELECT count(*) as maternal_deaths,
    residence_state||residence_municipality||residence_locality as locality_code,
    extract(year from date_of_death) as death_year
FROM maternal_mortality_data
GROUP BY locality_code, death_year
;""", db)

locality_info = pd.read_sql("""
SELECT cve_loc as locality_code,
    nom_loc as locality_name,
    nom_ent as state_name,
    cve_ent as state_code,
    nom_mun as municipality_name,
    cve_mun as municipality_code,
    pob_tot as population,
    VPHA as occupied_dwellings,
    ana_10 as pct_no_read,
    SNPRIM10 as pct_no_pri_ed,
    SNEXC10 as pct_no_toilet,
    SNEE10 as pct_no_elec,
    SAGUE10 as pct_no_water,
    PROM_OC10 as ppl_in_room,
    PISOTER10 as pct_earth_floor,
    SNREF10 as pct_no_fridge,
    IM_2010 as margination_index,
    GM_2010 as margination_degree,
    CON_UBIC as urbanicity
FROM localities
;""", db)
```

```python
### some data munging
births.rename(columns={'birth_year': 'year'}, inplace=True)
maternal_deaths.rename(columns={'death_year':'year'}, inplace=True)

data = pd.merge(births, maternal_deaths, 
                how='outer',
                on = ['locality_code', 'year'])
data.maternal_deaths.fillna(0, inplace=True)

data = pd.merge(data, locality_info, how='left', on='locality_code')
data = data[(data.year >= 2008) & (data.year <= 2012)]

def determine_urbanicity(con_ubic):
    if con_ubic == 0:
        return 'urban'
    elif con_ubic in [1,2,3,4]:
        return 'rural'
    else:
        return 'unknown'
data['urban_rural'] = data['urbanicity'].apply(lambda row: determine_urbanicity(row))
```

## Analysis of Urban vs. Rural Localities over time
- Looks like urban and rural areas have decreased in aggregate similarly, 20% between 2008 and 2012
- Large drop from 2009 to 2010 but a slowdown thereafter in both areas
- In agggregate, rural MMRs are ~2x higher than urban MMRs

```python
### Find MMR by urbanicity_type by year
urbanicity_by_year = data[data.urban_rural != 'unknown'].groupby(['year', 'urban_rural']).aggregate({'births':sum, 'maternal_deaths': sum})
urbanicity_by_year['mmr'] = urbanicity_by_year.maternal_deaths / urbanicity_by_year.births * 100000
urbanicity_by_year_mmr = urbanicity_by_year[['mmr']]
urbanicity_by_year_mmr = urbanicity_by_year_mmr.unstack('urban_rural').reset_index()
urbanicity_by_year_mmr.columns = urbanicity_by_year_mmr.columns.get_level_values(1)
urbanicity_by_year_mmr.rename(columns={'': 'year', 'rural': 'rural_mmr', 'urban': 'urban_mmr'}, inplace=True)
urbanicity_by_year_mmr['urban_mmr_change'] = float('NaN')
urbanicity_by_year_mmr['urban_pct_mmr_change'] = float('NaN')
urbanicity_by_year_mmr['rural_mmr_change'] = float('NaN')
urbanicity_by_year_mmr['rural_pct_mmr_change'] = float('NaN')

for year in urbanicity_by_year_mmr.year:
    try:
        urbanicity_by_year_mmr['urban_mmr_change'][urbanicity_by_year_mmr.year==year] = urbanicity_by_year_mmr['urban_mmr'][urbanicity_by_year_mmr.year==year].values[0] - urbanicity_by_year_mmr['urban_mmr'][urbanicity_by_year_mmr.year==year-1].values[0]
        urbanicity_by_year_mmr['rural_mmr_change'][urbanicity_by_year_mmr.year==year] = urbanicity_by_year_mmr['rural_mmr'][urbanicity_by_year_mmr.year==year].values[0] - urbanicity_by_year_mmr['rural_mmr'][urbanicity_by_year_mmr.year==year-1].values[0]
        urbanicity_by_year_mmr['urban_pct_mmr_change'][urbanicity_by_year_mmr.year==year] = (urbanicity_by_year_mmr['urban_mmr_change'][urbanicity_by_year_mmr.year==year].values[0] / urbanicity_by_year_mmr['urban_mmr'][urbanicity_by_year_mmr.year==year-1].values[0])*100
        urbanicity_by_year_mmr['rural_pct_mmr_change'][urbanicity_by_year_mmr.year==year] = (urbanicity_by_year_mmr['rural_mmr_change'][urbanicity_by_year_mmr.year==year].values[0] / urbanicity_by_year_mmr['rural_mmr'][urbanicity_by_year_mmr.year==year-1].values[0])*100
    except:
        pass
urbanicity_by_year_mmr
```

```python
# Plot percent change over time
plt.xticks(urbanicity_by_year_mmr.year, [str(int(i)) for i in urbanicity_by_year_mmr.year])
plt.plot(urbanicity_by_year_mmr.year, urbanicity_by_year_mmr.urban_pct_mmr_change, label='urban')
plt.plot(urbanicity_by_year_mmr.year, urbanicity_by_year_mmr.rural_pct_mmr_change, label='rural')
plt.legend()
plt.show()
```

```python
# Plot percent change over time
plt.xticks(urbanicity_by_year_mmr.year, [str(int(i)) for i in urbanicity_by_year_mmr.year])
plt.plot(urbanicity_by_year_mmr.year, urbanicity_by_year_mmr.urban_mmr, label='urban')
plt.plot(urbanicity_by_year_mmr.year, urbanicity_by_year_mmr.rural_mmr, label='rural')
plt.legend()
plt.show()
```

```python
## Find rural/urban differences by state over time
urbanicity_data = data[data.urban_rural != 'unknown'].groupby(['state_name', 'year', 'urban_rural'], as_index=False).aggregate({'births': sum, 'maternal_deaths': sum})
urbanicity_data['mmr'] = urbanicity_data.maternal_deaths/urbanicity_data.births * 100000

urbanicity_summary_data = pd.DataFrame()
for key, group in urbanicity_data.groupby(['state_name', 'urban_rural']):
    try:
        mmr_change = group.mmr[group.year==2012].values[0] - group.mmr[group.year==2008].values[0]
        pct_mmr_change = mmr_change / group.mmr[group.year==2008].values[0] * 100
        urbanicity_summary_data = urbanicity_summary_data.append({'state': key[0],
                                            'urbanicity': key[1],
                                            'mmr_change': mmr_change,
                                            'pct_mmr_change': pct_mmr_change,
                                            'mmr_2008': group.mmr[group.year==2008].values[0],
                                            'mmr_2012': group.mmr[group.year==2012].values[0],
                                            'births_2008': group.births[group.year==2008].values[0],
                                            'births_2012': group.births[group.year==2012].values[0],
                                            'maternal_deaths_2008': group.maternal_deaths[group.year==2008].values[0],
                                            'maternal_deaths_2012': group.maternal_deaths[group.year==2012].values[0]
                                            }, ignore_index=True)
    except:
        pass
urbanicity_summary_data['chi_sq'] = urbanicity_summary_data.apply(lambda row: proportion.proportions_chisquare([row.maternal_deaths_2008, row.maternal_deaths_2012], [row.births_2008, row.births_2012])[1], axis=1)
urbanicity_summary_data.sort(columns=['state', 'urbanicity'], inplace=True)
urbanicity_summary_data
```

```python
### Find urban/rural differences by state
urbanicity_data2 = urbanicity_data[urbanicity_data.year==2012]
urbanicity_summary_data2 = pd.DataFrame()
for key, group in urbanicity_data2.groupby(['state_name']):
    try:
        mmr_diff = group.mmr[group.urban_rural=='rural'].values[0] - group.mmr[group.urban_rural=='urban'].values[0]
        pct_mmr_diff = mmr_diff / group.mmr[group.urban_rural=='urban'].values[0] * 100
        urbanicity_summary_data2 = urbanicity_summary_data2.append({'state': key,
                                            'mmr_diff': mmr_diff,
                                            'pct_mmr_diff': pct_mmr_diff,
                                            'rural_mmr': group.mmr[group.urban_rural=='rural'].values[0],
                                            'urban_mmr': group.mmr[group.urban_rural=='urban'].values[0],
                    b                        'rural_births': group.births[group.urban_rural=='rural'].values[0],
                                            'urban_births': group.births[group.urban_rural=='urban'].values[0],
                                            'rural_maternal_deaths': group.maternal_deaths[group.urban_rural=='rural'].values[0],
                                            'urban_maternal_deaths': group.maternal_deaths[group.urban_rural=='urban'].values[0]
                                            }, ignore_index=True)
    except:
        pass
urbanicity_summary_data2['chi_sq'] = urbanicity_summary_data2.apply(lambda row: proportion.proportions_chisquare([row.urban_maternal_deaths, row.rural_maternal_deaths], [row.urban_births, row.rural_births])[1], axis=1)
urbanicity_summary_data2.sort(columns=['state'], inplace=True)
urbanicity_summary_data2
```
